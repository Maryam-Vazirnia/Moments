# MOMENTS

<div style="text-align: right">
مومنتها در آمار و احتمالات اندازه‌گیری‌هایی هستند که اطلاعاتی اضافی در مورد توزیع یک متغیر تصادفی فراهم می‌کنند. 
</div>

<div style="text-align: right">
مومنت اول (میانگین):مومنت اول یک متغیر تصادفی $X$، معمولاً به عنوان میانگین آن شناخته می‌شود. برای یک متغیر تصادفی $X$ با توزیع احتمال $f(x)$:
</div>
$ E(X)=\int xf(x)dx$
$ E(X)=μ$
<div style="text-align: right">
مومنت دوم (واریانس): واریانس یک متغیر تصادفی $X$ نشان دهنده اندازه پراکنش داده‌ها در اطراف میانگین است. برای یک متغیر تصادفی $X$ با میانگین $μ$، واریانس به صورت زیر محاسبه می‌شود:
</div>
$ Var(X)=E[(x−μ)^2] = \int (x−μ)^2f(x)dx$
<div style="text-align: right">
مومنت سوم (اسکیو $Skewness$): اسکیو یا شیب نشان دهنده تقارن توزیع است. مثبت بودن اسکیو نشان می‌دهد که داده‌ها به سمت دماغه توزیع بیشتر متمایل هستند و منفی بودن آن نشان دهنده تمایل به دماغه معکوس است. اگر اسکیو داده‌ها صفر باشد، داده‌ها تقریباً توزیع گاوسی دارند. 
مثبت یا منفی بودن این مومنت نشان‌دهندهٔ تمایل داده‌ها به رفتار نامتقارن در اطراف میانگین است (میزان انحراف از تقارن). به عبارتی اگر مومنت سوم مثبت باشد، داده‌ها به سمت مقادیر بزرگتر از میانگین کشیده شده‌اند و اگر منفی باشد، به سمت مقادیر کوچکتر از میانگین کشیده شده‌اند.
هر چقدر اسکیو به سمت صفر نزدیک‌تر باشد، داده‌ها به توزیع گاوسی نزدیک‌ترند
</div>
$ E[(X−μ)^3]$
<div style="text-align: right"> 
مومنت چهارم (کورتوسیس$Kurtosis$): کورتوسیس نشان دهنده شیب دماغه توزیع است (فراوانی داده‌ها در دم‌های بیشینه و کمینه). مقدار مثبت کورتوسیس نشان دهنده حضور دماغه بیشتر و تیزتر است و مقدار منفی نشان دهنده دماغه پخته تر و پخی است. هر چه کورتوسیس به سمت 3 نزدیک‌تر باشد، داده‌ها به توزیع گاوسی نزدیک‌ترند. اگر این مقدار بزرگتر از 3 باشد، داده‌ها شیب بیشتری به سمت دم‌های بیشینه و کمینه دارند.
</div>
$ E[(X−μ)^4]$

<div style="text-align: right"> 
مومنت پنجم: نشان‌دهندهٔ تمایل داده‌ها به داشتن حفره یا پیک‌های شدید در میانگین است.
</div>

<div style="text-align: right"> 
مومنت ششم: نشان‌دهندهٔ تمایل داده‌ها به داشتن حفره یا پیک‌های بیشتر در مقایسه با توزیع گاوسی است
</div>

### Previous explanations in English

Moments in statistics and probability are measures that provide additional information about the distribution of a random variable.

First moment (mean): The first moment of a random variable $X$ is usually known as its mean. For a random variable $X$ with probability distribution $f(x)$:

$E(X)=\int xf(x)dx$

$E(X)=μ$

The second moment (variance): The variance of a random variable $X$ indicates the size of the data distribution around the mean. For a random variable $X$ with mean $μ$, the variance is calculated as follows:

$ Var(X)=E[(x−μ)^2] = \int (x−μ)^2f(x)dx$

The third moment (Skew $Skewness$): Skew or slope indicates the symmetry of the distribution. The positive value of the skew indicates that the data are more inclined towards the nose of the distribution, and the negative value indicates the tendency towards the reverse nose. If the skewness of the data is zero, then the data has an approximately Gaussian distribution.
The positive or negative nature of this moment indicates the tendency of the data to behave asymmetric around the average (the degree of deviation from symmetry). In other words, if the third moment is positive, the data are drawn towards values larger than the average, and if it is negative, they are drawn towards values smaller than the average.
The closer the skew is to zero, the closer the data is to a Gaussian distribution

$E[(X−μ)^3]$

The fourth moment (Kurtosis$Kurtosis$): Kurtosis indicates the slope of the nose of the distribution (abundance of data in the maximum and minimum tails). A positive kurtosis value indicates the presence of a sharper nose, and a negative value indicates a more mature nose. The closer the kurtosis is to 3, the closer the data is to a Gaussian distribution. If this value is greater than 3, the data is more skewed towards the maximum and minimum tails.

$E[(X−μ)^4]$

The fifth moment: indicates the tendency of the data to have holes or extreme peaks in the average.

The sixth moment: indicates the tendency of the data to have more holes or peaks compared to the Gaussian distribution.



```python
import numpy as np
from scipy.stats import moment
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
##################################### First Example

data = [1, 2, 3, 4]
mean = moment(data, moment=1)
print("mean:", mean  )
variance = moment(data, moment=2)
print("variance:", variance)
skewness = moment(data, moment=3)
print("skewness:", skewness)
kurtosis = moment(data, moment=4)
print("kurtosis:", kurtosis)
```

```python
######################################## Second Example      previous data-Normalize

data = np.array([1, 2, 3, 4])
mean = np.mean(data)
print("mean:", mean)
variance = np.var(data)
print("variance:", variance)
skewness = np.mean((data - mean) ** 3) / np.std(data) ** 3
print("skewness:", skewness)
kurtosis = np.mean((data - mean) ** 4) / np.std(data) ** 4
print("kurtosis:", kurtosis)
```

```python
######################################## Third Example    Comparison between two data

np.random.seed(0)
data1 = np.random.normal(loc=5, scale=2, size=1000)
data2 = np.random.uniform(0, 100, size=1000)

mean1 = np.mean(data1)
variance1 = np.var(data1)
skewness1 = np.mean((data1 - mean1) ** 3) / np.std(data1) ** 3
kurtosis1 = np.mean((data1 - mean1) ** 4) / np.std(data1) ** 4

mean2 = np.mean(data2)
variance2 = np.var(data2)
skewness2 = np.mean((data2 - mean2) ** 3) / np.std(data2) ** 3
kurtosis2 = np.mean((data2 - mean2) ** 4) / np.std(data2) ** 4

print("Data 1:")
print("mean1:", mean1)
print("variance1:", variance1)
print("skewness1:", skewness1)
print("kurtosis1:", kurtosis1)
print("\nData 2:")
print("mean2:", mean2)
print("variance2:", variance2)
print("skewness2:", skewness2)
print("kurtosis2:", kurtosis2)

plt.figure(figsize=(12, 6)) , plt.subplot(1, 2, 1)
plt.hist(data1, bins=30, density=True, alpha=0.6, color='b', label='Data 1')
x1 = np.linspace(min(data1), max(data1), 100)
pdf1 = norm.pdf(x1, mean1, np.sqrt(variance1))
plt.plot(x1, pdf1, 'k-', linewidth=2, label='Normal distribution') , plt.title('DATA 1') , plt.xlabel('Values') , plt.ylabel('Probability density')
plt.legend() , plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(data2, bins=30, density=True, alpha=0.6, color='r', label='Data 2')
x2 = np.linspace(min(data2), max(data2), 100)
pdf2 = norm.pdf(x2, mean2, np.sqrt(variance2))
plt.plot(x2, pdf2, 'k-', linewidth=2, label='Normal distribution') , plt.title('DATA 2') , plt.xlabel('Values') , plt.ylabel('Probability density')
plt.legend() , plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Function to plot data distribution and Gaussian distribution

def plot_distribution(data, title):
    mean = np.mean(data)
    variance = np.var(data)
    
    # Gaussian distribution corresponding to the data
    x = np.linspace(min(data), max(data), 100)
    pdf = norm.pdf(x, mean, np.sqrt(variance))
    
    plt.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Data')
    plt.plot(x, pdf, 'r-', linewidth=2, label='Gaussian distribution')
    plt.title(title)
    plt.xlabel('Values')
    plt.ylabel('Probability densityل')
    plt.legend()
    plt.grid(True)
    plt.show()

np.random.seed(0)
data = np.random.randn(100)
#data = np.random.uniform(1,100,10000)
#data = np.random.normal(loc=0, scale=1, size=10000)
#data = [-3,-2,-1,0,1,2,3,4,5,6]
#data =  [3, 5, 7, 9, 11,12,15,23,54,23,12,43]

plot_distribution(data, 'data distribution')
```

```python
######################################## This section is exclusively for calculations for my own data
######################################## QUOTAS Data

input_file_path = 'QUOTAS.csv'
df = pd.read_csv(input_file_path)
df = df.dropna(subset=['z', 'log_bh', 'log_lbol'])
masses = df['log_bh'].values
luminosity = df['log_lbol'].values

mass_mean = np.mean(masses)
mass_variance = np.var(masses)
mass_skewness = np.mean((masses - mass_mean) ** 3) / np.std(masses) ** 3
mass_kurtosis = np.mean((masses - mass_mean) ** 4) / np.std(masses) ** 4

luminosity_mean = np.mean(luminosity)
luminosity_variance = np.var(luminosity)
luminosity_skewness = np.mean((luminosity - luminosity_mean) ** 3) / np.std(luminosity) ** 3
luminosity_kurtosis = np.mean((luminosity - luminosity_mean) ** 4) / np.std(luminosity) ** 4

print("Mass Moments:")
print("Mean:", mass_mean)
print("Variance:", mass_variance)
print("Skewness:", mass_skewness)
print("Kurtosis:", mass_kurtosis)

print("\nLuminosity Moments:")
print("Mean:", luminosity_mean)
print("Variance:", luminosity_variance)
print("Skewness:", luminosity_skewness)
print("Kurtosis:", luminosity_kurtosis)

# Comparison with Gaussian distribution
alpha = 0.05
if (abs(mass_skewness) < alpha) and (abs(mass_kurtosis - 3) < alpha):
    print("\nMass data is approximately normally distributed.")
else:
    print("\nMass data is not approximately normally distributed.")

if (abs(luminosity_skewness) < alpha) and (abs(luminosity_kurtosis - 3) < alpha):
    print("Luminosity data is approximately normally distributed.")
else:
    print("Luminosity data is not approximately normally distributed.")
```
